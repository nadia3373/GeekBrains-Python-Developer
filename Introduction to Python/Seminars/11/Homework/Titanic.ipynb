```python
import pandas as pd
data = pd.read_csv('data.csv')
```

# Общая информация

С помощью метода .describe() получаем основную статистическую информацию о наборе данных. На этом этапе уже можно сказать, что самому молодому пассажиру на Титанике было 0.42 года, а самому пожилому – 80 лет.
Средняя выживаемость пассажиров составляет 38%.

```python
data.describe()
```

# Сравнение количества выживших по полу

```python
data.query('Survived == 1').groupby(['Sex']).agg({'PassengerId': 'count'})
```

Видно, что количество выживших женщин более чем в два раза превышает количество выживших мужчин, и это подтверждается историческими фактами – у женщин был приоритет при посадке в спасательные шлюпки.

# Общее количество пассажиров по классам

```python
data.groupby(['Pclass']).agg({'PassengerId': 'count'})
```

Пассажиров 3 класса на корабле было более, чем в 2.5 раза больше, чем пассажиров 1 или 2 класса, что объясняется доступностью билетов 3 класса и большим количеством небогатых мигрантов, стремившихся попасть в США в начале XX века.

# Статистика выживших пассажиров по классу

```python
data.query('Survived == 1').groupby(['Pclass']).agg({'PassengerId': 'count'})
```

Результаты по классам также совпадают с историческими фактами: наибольшее количество выживших – это пассажиры 1 класса, что было связано с расположением их кают на верхних ярусах корабля, возможностью беспрепятственного прохода к шлюпкам, а также поначалу приоритетной посадкой пасажиров 1 класса. Достаточно большое количество выживших пассажиров 3 класса объясняется подавляющим большинством пассажиров этого класса на корабле.

# Статистика выживших пассажиров по полу и классу

```python
data.query('Survived == 1').groupby(['Pclass', 'Sex']).agg({'PassengerId': 'count'})
```

Наибольшая разница между количеством выживших женщин и мужчин наблюдается среди пассажиров 2 класса. Самая многочисленная категория выживших – женщины из числа пассажиров 1 класса.

# Количество пассажиров по портам посадки

```python
data.groupby('Embarked').agg({'PassengerId': 'count'})
```

Наибольшее количество пассажиров отбыло из Саутгемптона.

# Статистика выживших пассажиров в зависимости от порта посадки

```python
data.query('Survived == 1').groupby('Embarked').agg({'PassengerId': 'count'})
```

Среди выживших пассажиров наибольшее количество также из Саутгемптона.
